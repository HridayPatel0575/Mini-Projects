```python
cols_to_cap = [
    'Salary',
    'Errors',
    'Assists',
    'PutOuts',
    'CWalks',
    'CRBI',
    'CHmRun',
    'HmRun',        
    'Runs' ,        
    'RBI'  ,       
    'Walks'   ,    
    'Years'    ,   
    'CAtBat'  ,
    'CHits',
    'CRuns'
    
]

for col in cols_to_cap:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    print(f"{col}: lower={lower_limit}, upper={upper_limit}")

    df[col] = df[col].clip(lower_limit, upper_limit)

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=['Salary']),df['Salary'],test_size=0.2,random_state=42)
```

```python
skew_values = x_train.select_dtypes(include=[np.number]).skew().sort_values(ascending=False)
print(skew_values)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
alphas = 10**np.linspace(10,-2,100)*0.00001
ridge = Ridge()
coeffs = []
```

```python
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(x_train_sc,y_train)
    coeffs.append(ridge.coef_)
```

```python
from sklearn.metrics import mean_squared_error

train_mse = []
test_mse = []

for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(x_train_sc, y_train)
    
    train_mse.append(mean_squared_error(y_train, ridge.predict(x_train_sc)))
    test_mse.append(mean_squared_error(y_test, ridge.predict(x_test_sc)))
```

```python
plt.figure(figsize=(8,5))
plt.plot(alphas, train_mse, label='Train MSE')
plt.plot(alphas, test_mse, label='Test MSE')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.legend()
plt.title('Bias–Variance Tradeoff (Ridge)')
plt.show()
```

```python
best_alpha = alphas[np.argmin(test_mse)]
print("Best alpha:", best_alpha)
```

```python
ridge_final = Ridge(alpha=best_alpha)
ridge_final.fit(x_train_sc, y_train)
```

```python
y_pred = ridge_final.predict(x_test_sc)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
print(mean_squared_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100)

rf.fit(x_train_sc,y_train)

y_pred_rf=rf.predict(x_test_sc)

print(mean_squared_error(y_test,y_pred_rf))
print(root_mean_squared_error(y_test,y_pred_rf))
print(r2_score(y_test,y_pred_rf))
```

```python
model_2 = Sequential()
model_2.add(Dense(128, input_dim=x_train_sc.shape[1], activation="relu"))
model_2.add(Dropout(0.2))
model_2.add(Dense(512, activation="relu"))
model_2.add(Dropout(0.2))
model_2.add(Dense(1, activation="linear"))

adam = Adam(learning_rate=0.01)
model_2.compile(loss='mse', optimizer=adam, metrics=['mse'])
```

```python
history = model_2.fit(
    x_train_sc, y_train,
    epochs=500,
    validation_data=(x_test_sc, y_test),
    verbose=True
)
```

```python
y_pred_nn = model_2.predict(x_test_sc).ravel()

rmse_nn = np.sqrt(mean_squared_error(y_test, y_pred_nn))
r2_nn = r2_score(y_test, y_pred_nn)

print("Neural Network RMSE:", rmse_nn)
print("Neural Network R2  :", r2_nn)
```

```python
from sklearn.linear_model import Lasso

alphas = 10**np.linspace(10, -2, 100) * 0.00001
lasso = Lasso(max_iter=100000)
coeffs = []
```

```python
for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(x_train_sc, y_train)
    coeffs.append(lasso.coef_)
```

```python
train_mse = []
test_mse = []

for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(x_train_sc, y_train)
    
    train_mse.append(mean_squared_error(y_train, lasso.predict(x_train_sc)))
    test_mse.append(mean_squared_error(y_test, lasso.predict(x_test_sc)))
```

```python
plt.figure(figsize=(8,5))
plt.plot(alphas, train_mse, label='Train MSE')
plt.plot(alphas, test_mse, label='Test MSE')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.legend()
plt.title('Bias–Variance Tradeoff (Lasso)')
plt.show()
```

```python
best_alpha = alphas[np.argmin(test_mse)]
print("Best alpha:", best_alpha)
```

```python
lasso_final = Lasso(alpha=best_alpha, max_iter=100000)
lasso_final.fit(x_train_sc, y_train)

y_pred = lasso_final.predict(x_test_sc)
```

```python
print(mean_squared_error(y_test, y_pred))
print(root_mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
